```python
import os
import shutil
import random

source_dir = "train"   # ana klasör
output_dir = "output"  # Yeni oluşacak düzenli klasör
train_ratio = 0.8
val_ratio   = 0.1      # Test de geri kalan %10 olacak
# ==================

# Kaynak alt klasör yolları
src_images_dir = os.path.join(source_dir, "images")
src_labels_dir = os.path.join(source_dir, "labels")

# Hedef Klasörleri oluştur (Train/Val/Test)
train_img = os.path.join(output_dir, "train/images")
train_lbl = os.path.join(output_dir, "train/labels")
val_img   = os.path.join(output_dir, "val/images")
val_lbl   = os.path.join(output_dir, "val/labels")
test_img  = os.path.join(output_dir, "test/images")
test_lbl  = os.path.join(output_dir, "test/labels")

for d in [train_img, train_lbl, val_img, val_lbl, test_img, test_lbl]:
    os.makedirs(d, exist_ok=True)

# Görüntüleri "images" klasöründen oku
images = [f for f in os.listdir(src_images_dir) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
random.shuffle(images)

# Split (Bölme) hesaplama
total = len(images)
train_count = int(total * train_ratio)
val_count   = int(total * val_ratio)

train_files = images[:train_count]
val_files   = images[train_count : train_count+val_count]
test_files  = images[train_count+val_count :]

def move_files(file_list, img_dest, lbl_dest):
    for img_name in file_list:
        # Resim yolu
        img_src_path = os.path.join(src_images_dir, img_name)
        
        # Etiket adı ve yolu (Resmin adını al, sonuna .txt ekle)
        lbl_name = os.path.splitext(img_name)[0] + ".txt"
        lbl_src_path = os.path.join(src_labels_dir, lbl_name)

        # Resmi kopyala
        shutil.copy(img_src_path, img_dest)

        # Etiket dosyası varsa kopyala
        if os.path.exists(lbl_src_path):
            shutil.copy(lbl_src_path, lbl_dest)
        else:
            print(f"Uyarı: Etiket dosyası bulunamadı → {lbl_name}")

# Kopyalama işlemini başlat
print("Dosyalar ayrıştırılıyor, lütfen bekleyin...")
move_files(train_files, train_img, train_lbl)
move_files(val_files,   val_img,   val_lbl)
move_files(test_files,  test_img,  test_lbl)

# Rapor
print("\n Dataset Split Tamamlandı!")
print(f"Toplam görüntü: {total}")
print(f"Train: {len(train_files)} | Val: {len(val_files)} | Test: {len(test_files)}")
print(f"\nYeni dataset dizini: {output_dir}/\n")
```
